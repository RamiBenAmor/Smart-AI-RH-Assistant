```python
# ‚úÖ 1. Imports & Pr√©paration
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from collections import Counter
import numpy as np
import torch
from transformers import BertTokenizer, BertForSequenceClassification, TrainingArguments, DataCollatorWithPadding, Trainer
from datasets import Dataset
import torch.nn as nn
import evaluate
from sklearn.metrics import classification_report

# ‚úÖ 2. Donn√©es
df = df[['Feature', 'Category']]
df['Feature'] = df['Feature'].astype(str).fillna('')
le = LabelEncoder()
df['label'] = le.fit_transform(df['Category'])
df['labels'] = df['label']
num_labels = len(le.classes_)

# ‚úÖ 3. Tokenizer & Dataset
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
def tokenize(batch):
    return tokenizer(batch['Feature'], padding=True, truncation=True, max_length=512)
dataset = Dataset.from_pandas(df)
dataset = dataset.map(tokenize, batched=True)

# ‚úÖ 4. Split stratifi√©
train_idx, val_idx = train_test_split(range(len(dataset)), test_size=0.2, stratify=df['label'], random_state=42)
train_dataset = dataset.select(train_idx)
val_dataset = dataset.select(val_idx)

# ‚úÖ 5. Mod√®le
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)

# ‚úÖ 6. Poids de classes
class_counts = Counter(df["label"])
freqs = np.array([class_counts[i] for i in range(num_labels)])
inv_freqs = 1.0 / freqs
class_weights = inv_freqs * (num_labels / inv_freqs.sum())
class_weights = torch.tensor(class_weights, dtype=torch.float)

# ‚úÖ 7. TrainingArguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=10,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="f1_weighted",
    report_to="none"
)

# ‚úÖ 8. Metrics
f1_metric = evaluate.load("f1")
acc_metric = evaluate.load("accuracy")
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    f1 = f1_metric.compute(predictions=preds, references=labels, average="weighted")["f1"]
    acc = acc_metric.compute(predictions=preds, references=labels)["accuracy"]
    return {"f1_weighted": f1, "accuracy": acc}

# ‚úÖ 9. Data collator
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None): # Added num_items_in_batch
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = nn.CrossEntropyLoss(weight=class_weights.to(model.device))
        loss = loss_fct(logits, labels)
        return (loss, outputs) if return_outputs else loss
from transformers import EarlyStoppingCallback

# ‚úÖ 11. Entra√Ænement
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],
)
trainer.train()

# ‚úÖ 12. √âvaluation d√©taill√©e
predictions = trainer.predict(val_dataset)
y_true = predictions.label_ids
y_pred = np.argmax(predictions.predictions, axis=-1)
print(classification_report(y_true, y_pred, target_names=le.classes_))
```

```python
#correct one
#Toujours faire le split train/test en premier, puis appliquer l‚Äô√©quilibrage uniquement sur le jeu d‚Äôentra√Ænement.

import pandas as pd
from sklearn.utils import resample
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report, f1_score
from sklearn.naive_bayes import MultinomialNB
import joblib

# Exemple avec base_models d√©j√† d√©fini et entra√Æn√©
trained_pipelines = {}
# Supposons que df est ton DataFrame original avec colonnes 'Feature' (texte) et 'Category' (label)
# Exemple:
# df = pd.read_csv("ton_dataset.csv")

# 1. √âquilibrage des classes avec sur/sous-√©chantillonnage selon la m√©diane
target_count = int(df['Category'].value_counts().median())
balanced_data = []

for category in df['Category'].unique():
    category_data = df[df['Category'] == category]
    current_count = len(category_data)
    if current_count < target_count:
        # Sur-√©chantillonnage
        balanced_category_data = resample(
            category_data,
            replace=True,
            n_samples=target_count,
            random_state=42
        )
    elif current_count > target_count:
        # Sous-√©chantillonnage
        balanced_category_data = resample(
            category_data,
            replace=False,
            n_samples=target_count,
            random_state=42
        )
    else:
        balanced_category_data = category_data
    balanced_data.append(balanced_category_data)

balanced_df = pd.concat(balanced_data).sample(frac=1, random_state=42).reset_index(drop=True)

# S‚Äôassurer que la colonne 'Feature' est bien du texte
balanced_df['Feature'] = balanced_df['Feature'].astype(str).fillna('')

# 2. Encodage des labels
le = LabelEncoder()
balanced_df['label'] = le.fit_transform(balanced_df['Category'])

# 3. Split stratifi√© train/test sur les donn√©es brutes (texte + labels)
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    balanced_df['Feature'], balanced_df['label'],
    stratify=balanced_df['label'], test_size=0.2, random_state=42
)

# 4. D√©finition des mod√®les classiques
base_models = {
    "LogReg": LogisticRegression(max_iter=1000, class_weight="balanced"),
    "RandomForest": RandomForestClassifier(),
    "SVC": SVC(class_weight="balanced", kernel="linear"),
    "KNN": KNeighborsClassifier(),
    "OneVsRest(KNN)": OneVsRestClassifier(KNeighborsClassifier())
}
L=[]
# 5. Entra√Ænement et √©valuation avec pipeline int√©grant TF-IDF
for name, clf in base_models.items():
    pipe = Pipeline([
        ('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1, 2))),
        ('clf', clf)
    ])
    pipe.fit(X_train_raw, y_train)
    y_pred = pipe.predict(X_test_raw)
    f1 = f1_score(y_test, y_pred, average="weighted")
    print(f"Model: {name}")
    print(f"‚úÖ F1-score pond√©r√© : {f1:.4f}")
    L.append(f1)
    trained_pipelines[name] = pipe
    joblib.dump(pipe, f"{name}_model.joblib")
    print(classification_report(y_test, y_pred, target_names=le.classes_))
    print("\n")


# 6. MultinomialNB sans SMOTE, aussi avec pipeline
nb_pipe = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1, 2))),
    ('clf', MultinomialNB())
])
nb_pipe.fit(X_train_raw, y_train)
y_nb = nb_pipe.predict(X_test_raw)
print("===== MultinomialNB (sans SMOTE) =====")
print(f"‚úÖ F1-score pond√©r√© : {f1_score(y_test, y_nb, average='weighted'):.4f}")
print(classification_report(y_test, y_nb, target_names=le.classes_))
trained_pipelines["MultinomialNB"] = nb_pipe
joblib.dump(nb_pipe, "MultinomialNB_model.joblib")

print("Mod√®les sauvegard√©s.")
```
